```python
x = np.linspace(1,5,1000)
y1 = np.sin(x)
y2 = np.exp(0.5*x)
y3 = np.exp(-np.power(x - -1, 2.) / (2 * np.power(1, 2.)))
y4 = np.log10(x)
y5 = np.arctan(x)

def my_plot(x,y,xlabel='',ylabel='',title='',color='#1f77b4'):
    fig,ax = plt.subplots()
    ax.plot(x,y,color)
    ax.set(xlabel=xlabel,ylabel=ylabel,title=title)

my_plot(x,y1)
```

```python
my_plot(x,y2,'Oś X',"Oś Y","Jakiś tytuł",'r')
```

```python
my_plot(x,y3,'Oś X',"Oś Y",":))",'g')
```

```python
my_plot(x,y4,'Oś X',title='fdhsfisdgh',color='purple')
```

```python
my_plot(x,y5,'Oś X',color='y')
```

```python
x_i=np.array([-2.25, -1.5, -0.75, 0, 0.75, 1.5, 2.25, 3])
y_i=np.array([0,0,-0.931596,0,0.931596, 14.1014, 13, 14])

lag_base= lambda i,x : np.prod(x-np.delete(x_i,i))/np.prod(x_i[i]-np.delete(x_i,i))
Y=[]
X=np.linspace(-1.499,1.499,100)
for x in np.nditer(X):
    L=np.array([lag_base(0,x),lag_base(1,x),lag_base(2,x),lag_base(3,x),lag_base( 4,x),lag_base(5,x),lag_base(6,x),lag_base(7,x)])
    Y.append(y_i@L.transpose())
Y=np.array(Y)
```

```python
plt.plot(X,Y)
plt.scatter(x_i,y_i)
```
