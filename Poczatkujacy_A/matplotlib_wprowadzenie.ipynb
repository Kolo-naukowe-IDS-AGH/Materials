```python
x = np.linspace(0,10,11)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.scatter(x,y,s=100,c='r',alpha=0.5,marker='*') # jest mnóstwo parametrów pozwalających dostosowywać wykresy
plt.xlabel("Oś X")
plt.ylabel("Oś Y")
plt.title("Przykładowy wykres")
plt.grid()
plt.legend(['y1','y2'])
#plt.xlim(-2,12)
#plt.ylim(-2,2)
plt.show()
```

```python
# Create a Figure and an Axes with plt.subplots
fig, ax = plt.subplots()
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(2,2)
#fig.set_size_inches(15,10)
x = np.linspace(0,10,1000)
y = 2*x**2 - 12*x + 2
ax[0,0].plot(x,y,'r')
ax[0,1].plot(x,y,'g')
ax[1,0].plot(x,y,'y')
ax[1,1].plot(x,y,'b')
plt.show()
```

#### Kilka przykładów tego, co potrafi matplotlib:

```python
x = np.linspace(1,5,1000)
y1 = np.sin(x)
y2 = np.exp(0.5*x)
y3 = np.exp(-np.power(x - -1, 2.) / (2 * np.power(1, 2.)))
y4 = np.log10(x)
y5 = np.arctan(x)

def my_plot(x,y,xlabel='',ylabel='',title='',color='#1f77b4'):
    fig,ax = plt.subplots()
    ax.plot(x,y,color)
    ax.set(xlabel=xlabel,ylabel=ylabel,title=title)

my_plot(x,y1)
```

```python
my_plot(x,y2,'Oś X',"Oś Y","Jakiś tytuł",'r')
```

```python
my_plot(x,y3,'Oś X',"Oś Y",":))",'g')
```

```python
my_plot(x,y4,'Oś X',title='fdhsfisdgh',color='purple')
```

```python
my_plot(x,y5,'Oś X',color='y')
```
