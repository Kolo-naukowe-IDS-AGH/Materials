# <center>NumPy</center>
<center>(short for Numerical Python)

NumPy (short for Numerical Python) provides an efficient interface to store and operate on dense data buffers.
In some ways, NumPy arrays are like Python's built-in list type, but NumPy arrays provide much more efficient
 storage and data operations as the arrays grow larger in size. NumPy arrays form the core of nearly the 
entire ecosystem of data science tools in Python, so time spent learning to use NumPy effectively will
 be valuable no matter what aspect of data science interests you.

```python
import numpy as np
import matplotlib.pyplot as plt
print(np.__version__)
```

```python
a = np.array([1, 2, 3])
b = np.array([[1,2,3],[4,5,6]]) 
print(type(a))
print(a)
print(a.shape)
print(b)
print(b.shape)
```

```python
print(np.sum(b))
print(np.prod(b))
print(np.mean(b))
print(np.median(b))
print(np.std(b))
print(np.var(b))
print(np.max(b))
print(np.min(b))
```

```python
c = np.zeros(5)
d = np.zeros((5,5))
print(c,"\n")
print(d)
```

```python
e = np.eye(6)
print(e,"\n")
f = np.ones((3,6))
print(f)
```

```python
g = np.arange(1,20,3) # return evenly spaced values within a given interval
print(g,'\n')
h = np.linspace(0,10,12) # returns num (12) evenly spaced samples, calculated over the interval
print(h)
```

```python
h_reshaped = h.reshape(3,4)
#h_reshaped = h.reshape(3,7) #wyskoczy błąd, bo nie da się przekształcić macierzy 1x12 na 3x7
print(h_reshaped,"\n")
h_res_T = np.transpose(h_reshaped) # można też h_reshaped.T 
print(h_res_T)
```

```python
arr1 = np.array([[5,4,3],[10,5,0]])
arr2 = np.array([[1,2],[3,4],[5,6]]) 
print(np.dot(arr1,arr2),'\n') #mnożenie macierzowe
print(arr1 @ arr2) #drugi sposób
```

```python
np.random.seed(123)
x = np.random.rand(2,3)
y = np.random.randn(2,3)
z = np.random.randint(2,10,size=(2,3))
#dużo możliwości: https://docs.scipy.org/doc/numpy-1.15.0/reference/routines.random.html
print(x,"\n")
print(y,"\n")
print(z,"\n")
```

```python
print(z*y,"\n")
print(x+z,"\n")
print(z/y)
```

```python
array = np.random.randint(10,size=(3,5))
print("Array:\n {}\n".format(array))
print(array[0],'\n')
print(array[1:],'\n')
print(array[2,3])
print(array[2][3])
x = np.array([1,2,3,4,5,6])
print(x<4)
print(x[x<4])
```
